### 2a. Using qiskit, implement in QASM the four pulse sequence given in Eq.(1). ###

This function below implements the BB1 sequence when $\theta = \pi$. It also takes in a $\epsilon$ variable as we manual consider the error on the QASM simulator

```python
def get_circuit(epsilon):
    qc = QuantumCircuit(1,1)
   
    # M0 gate
    qc.rz(0,0)
    qc.rx(np.pi + epsilon,0)
    qc.rz(0,0)
    
    # M_phi1 gate
    qc.rz(-np.arccos(-1/4),0)
    qc.rx(np.pi + epsilon,0)
    qc.rz(np.arccos(-1/4),0)
    
    #M_phi2 gate
    qc.rz(-3*np.arccos(-1/4),0)
    qc.rx(2 *np.pi + 2* epsilon,0)
    qc.rz(3*np.arccos(-1/4),0)
    
    # M_phi1 gate
    qc.rz(-np.arccos(-1/4),0)
    qc.rx(np.pi + epsilon,0)
    qc.rz(np.arccos(-1/4),0)

    qc.measure(0,0)
    return qc
```

### 2b. Think about how to test your implementation, implement this test (in simulation), and convince group memebers that your implementation works successfully. ###

To test my implementation, I will arrange a range of $\epsilon$ from $(-\pi, \pi)$, and simulate the BB1 composite pulse under the effect of different $\epsilon$. I can then plot my the probability of bit flip at different $\epsilon$ obtained from the simulator, and compare against the manually calculated from of p obtained in 1c. The goal is to see if our simulated result matches our hand-calculated result

```python
epsilon = np.arange(-np.pi,np.pi,0.001)
n_shots = 5000
simulated_p=[]

for i in range(len(epsilon)):
    qc = get_circuit(epsilon[i])
    job = qasm_simulator_backend.run(qc, shots=n_shots)
    result = job.result()

    count0 = 0
    count1 = 0
    if ('0' in result.get_counts()):
        count0 = result.get_counts()['0']
    if ('1' in result.get_counts()):
        count1 = result.get_counts()['1']
    simulated_p.append(count1/n_shots)
```

```python
plt.figure(figsize=(12,6),dpi=600)
plt.plot(epsilon,calculated_p,label='Calculated p')
plt.plot(epsilon,simulated_p,label='Simulated p')
plt.xlim(-np.pi,np.pi)
#plt.ylim(0,1.1)
plt.legend(loc='upper right',fontsize=14)
plt.show()
```

We would need a way to make sure we correct convert the Rz/Rx gates to IonQ native gates. To do this, I will use the QuantumMatrix class provided in this link: https://github.com/qiskit-community/qiskit-community-tutorials/blob/master/terra/qis_adv/Clifford_Group.ipynb. It offers a good way to visualize the matrix and also takes care of the phase and coefficients thus allowing more direct comparison.

```python
class QuantumMatrix():
    def __init__(self, m, coeff = 1):
        self.matrix = sympy.Matrix(m)
        self.coefficient = coeff
        self.canonize()
        
    def canonize(self):
        a = next((x for x in self.matrix if x != 0), None)
        if a is not None: #zero vector
            for i,j in itertools.product([0,1], [0,1]):
                self.matrix[i,j] = sympy.simplify(self.matrix[i,j] / a)
        self.coefficient = sympy.simplify(self.coefficient * a) 
            
    def __str__(self):
        coeff_string = ""
        if self.coefficient != 1:
            coeff_string = "{} * ".format(self.coefficient)
        return "{}[[{}, {}], [{}, {}]]".format(coeff_string,self.matrix[0], self.matrix[1], self.matrix[2], self.matrix[3])
    
    def __mul__(self, rhs):
        return QuantumMatrix(self.matrix * rhs.matrix, self.coefficient * rhs.coefficient)
    
    def __add__(self, rhs):
        temp_rhs_matrix = sympy.Matrix([[1,0],[0,1]])
        for i,j in itertools.product([0,1], [0,1]):
                 temp_rhs_matrix[i,j] = sympy.simplify((rhs.matrix[i,j] * self.coefficient) / rhs.coefficient)
        return QuantumMatrix(self.matrix + temp_rhs_matrix, self.coefficient * 1/sympy.sqrt(2))
    
    def __sub__(self, rhs):
        return self + QuantumMatrix(rhs.matrix, rhs.coefficient * -1)
    
    def __eq__(self, rhs):
        return (self.matrix == rhs.matrix and self.coefficient == rhs.coefficient)
    
    def equiv(self, rhs):
        return (self.matrix == rhs.matrix)
    
    def __iter__(self):
        for x in self.matrix:
             yield x
```

```python
#Define Rotation matrices as a function of theta, e.g. Rx(theta), etc.
Rx = lambda theta : QuantumMatrix([[    sympy.cos(theta/2),-1j*sympy.sin(theta/2)],
                              [-1j*sympy.sin(theta/2),    sympy.cos(theta/2)]])
Ry = lambda theta : QuantumMatrix([[    sympy.cos(theta/2),   -sympy.sin(theta/2)],
                              [    sympy.sin(theta/2),    sympy.cos(theta/2)]])
Rz = lambda theta : QuantumMatrix([[sympy.exp(-sympy.I*theta/2),                0.0],
                              [                0.0, sympy.exp(sympy.I*theta/2)]])
```

```python
#Define IonQ native gates as function of phi
GPI = lambda phi : QuantumMatrix( [[0,sympy.exp(-sympy.I * phi)], 
                                   [sympy.exp(sympy.I * phi),0]])


GPI2 = lambda phi : QuantumMatrix( [[1, -sympy.I * sympy.exp(-sympy.I * phi)], 
                                     [ -sympy.I * sympy.exp(sympy.I * phi),1]], 1/sympy.sqrt(2))
```
